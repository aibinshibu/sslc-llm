```python
import pandas as pd
import numpy as np
import math
import json
import os
import ipywidgets as widgets
from IPython.display import display, clear_output
from sentence_transformers import SentenceTransformer, util
import torch

# ==========================================
# 1. CONFIGURATION
# ==========================================
FILE_NAME = '/content/geoode.csv'
VECTOR_COL = 'Activity_Vector'
MODEL_NAME = 'all-mpnet-base-v2'  # Or 'all-MiniLM-L6-v2' for speed

print(f"🚀 Initializing MSME Intelligence Engine...")
print(f"📊 Data Source: {FILE_NAME}")
print(f"🤖 AI Model: {MODEL_NAME}")

# ==========================================
# 2. LOAD OR INITIALIZE AI MODEL
# ==========================================
print(f"\n⏳ Loading {MODEL_NAME}...")
ai_model = SentenceTransformer(MODEL_NAME)
print(f"✅ AI Model Ready")

# ==========================================
# 3. SMART DATA LOADING WITH VECTOR CACHE
# ==========================================
def load_data_with_vectors():
    """
    Intelligent loading:
    - If vectors exist in CSV → Load instantly
    - If missing → Compute, save, then load
    """
    try:
        df = pd.read_csv(FILE_NAME)
        df.columns = df.columns.str.strip()

        # Standardize column names
        rename_map = {}
        for col in df.columns:
            if col.lower() in ['latitude', 'lat']: rename_map[col] = 'latitude'
            elif col.lower() in ['longitude', 'lon', 'lng']: rename_map[col] = 'longitude'
            elif col.lower() in ['activities', 'activity', 'sector']: rename_map[col] = 'Activities'
            elif col.lower() in ['enterprisename', 'name', 'enterprise_name']: rename_map[col] = 'EnterpriseName'
            elif col.lower() in ['communicationaddress', 'address']: rename_map[col] = 'CommunicationAddress'

        df.rename(columns=rename_map, inplace=True)
        df['Activities'] = df['Activities'].fillna("Unknown").astype(str)

        # CHECK: Do vectors already exist?
        if VECTOR_COL in df.columns:
            print(f"💾 Found pre-computed vectors! Loading instantly...")

            # Parse JSON vectors back to numpy arrays
            def parse_vector(vec_str):
                try:
                    return np.array(json.loads(vec_str), dtype=np.float32)
                except:
                    return None

            df['vector_parsed'] = df[VECTOR_COL].apply(parse_vector)

            # Handle any corrupted rows
            valid_vectors = df['vector_parsed'].notna().sum()
            if valid_vectors < len(df):
                print(f"⚠️ {len(df) - valid_vectors} records have missing/corrupt vectors. Recomputing those...")
                missing_idx = df[df['vector_parsed'].isna()].index
                new_vectors = ai_model.encode(df.loc[missing_idx, 'Activities'].tolist(), convert_to_numpy=True)

                for i, idx in enumerate(missing_idx):
                    df.at[idx, 'vector_parsed'] = new_vectors[i]
                    df.at[idx, VECTOR_COL] = json.dumps(new_vectors[i].tolist())

                # Save updated CSV
                df.to_csv(FILE_NAME, index=False)
                print(f"💾 Updated CSV with repaired vectors")

            # Convert to tensor for fast similarity search
            activity_embeddings = torch.tensor(np.stack(df['vector_parsed'].values), dtype=torch.float32)
            print(f"✅ Loaded {len(df)} records with vectors in ~0.5 seconds!")

        else:
            # FIRST TIME: Compute all vectors
            print(f"🔮 No vectors found. Computing embeddings for {len(df)} records...")
            print(f"⏳ This takes ~{len(df) * 0.002:.0f} seconds (one-time only)...")

            # Batch encoding for memory efficiency
            batch_size = 1000
            all_vectors = []

            for i in range(0, len(df), batch_size):
                batch = df['Activities'].iloc[i:i+batch_size].tolist()
                batch_vectors = ai_model.encode(batch, convert_to_numpy=True, show_progress_bar=False)
                all_vectors.extend(batch_vectors)
                print(f"  Processed {min(i+batch_size, len(df))}/{len(df)}...")

            # Store as JSON strings in CSV
            df[VECTOR_COL] = [json.dumps(v.tolist()) for v in all_vectors]
            activity_embeddings = torch.tensor(all_vectors, dtype=torch.float32)

            # Save enriched CSV
            df.to_csv(FILE_NAME, index=False)
            print(f"💾 Saved vectors to '{VECTOR_COL}' column in {FILE_NAME}")
            print(f"✅ Future loads will be instant!")

        return df, activity_embeddings

    except FileNotFoundError:
        print(f"❌ ERROR: '{FILE_NAME}' not found.")
        empty_df = pd.DataFrame(columns=['EnterpriseName', 'CommunicationAddress', 'Activities', 'latitude', 'longitude', VECTOR_COL])
        return empty_df, torch.tensor([])

# Execute smart loading
df, activity_embeddings = load_data_with_vectors()

# ==========================================
# 4. BUSINESS LOGIC ENGINE (Same as before)
# ==========================================
BUSINESS_CATEGORIES = {
    'retail': {
        'keywords': ['shop', 'store', 'retail', 'dealer', 'seller', 'sales', 'boutique', 'mart', 'showroom', 'reseller'],
        'anti_keywords': ['repair', 'service', 'mending', 'manufacturing', 'mfg', 'factory', 'industry', 'works', 'fabrication', 'production']
    },
    'service': {
        'keywords': ['repair', 'service', 'fixing', 'mending', 'alteration', 'maintenance', 'servicing', 'cobbler'],
        'anti_keywords': ['manufacturing', 'factory', 'mfg', 'production', 'industry', 'mill', 'retail', 'store', 'seller']
    },
    'manufacturing': {
        'keywords': ['manufacturing', 'mfg', 'factory', 'industry', 'production', 'mill', 'works', 'fabrication', 'maker', 'unit'],
        'anti_keywords': ['repair', 'service', 'retail', 'shop', 'store', 'seller', 'dealer']
    }
}

def classify_business_intent(query):
    query_lower = query.lower()
    scores = {}

    for category, data in BUSINESS_CATEGORIES.items():
        score = sum(2 for keyword in data['keywords'] if keyword in query_lower)
        penalty = sum(1 for anti in data['anti_keywords'] if anti in query_lower)
        final_score = score - penalty
        if final_score > 0:
            scores[category] = final_score

    if not scores:
        return 'general', 0.0

    best_category = max(scores, key=scores.get)
    return best_category, scores[best_category]

def haversine(lat1, lon1, lat2, lon2):
    R = 6371.0
    phi1, phi2 = np.radians(lat1), np.radians(lat2)
    d_phi = np.radians(lat2 - lat1)
    d_lambda = np.radians(lon2 - lon1)
    a = np.sin(d_phi/2)**2 + np.cos(phi1) * np.cos(phi2) * np.sin(d_lambda/2)**2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1-a))
    return R * c

def calculate_competitor_density(target_lat, target_lon, search_query, radius_km=2.0, ai_threshold=0.45):
    if df.empty or len(activity_embeddings) == 0:
        return 0, 0.0, pd.DataFrame(), pd.DataFrame()

    intent_category, _ = classify_business_intent(search_query)

    # ⚡ FAST: Only encode the query, not the whole database!
    query_embedding = ai_model.encode(search_query, convert_to_tensor=True)

    # ⚡ FAST: Pre-computed vectors loaded from CSV
    cosine_scores = util.cos_sim(query_embedding, activity_embeddings)[0]

    matching_indices = np.where(cosine_scores.cpu().numpy() >= ai_threshold)[0]
    potential_matches = df.iloc[matching_indices].copy()
    potential_matches['AI_Raw_Score'] = cosine_scores[matching_indices].cpu().numpy()

    # Strict filtering (same logic as before)
    filtered_matches = []

    for _, row in potential_matches.iterrows():
        activity_text = str(row['Activities']).lower()
        ai_score = row['AI_Raw_Score']

        if intent_category != 'general':
            rules = BUSINESS_CATEGORIES[intent_category]
            has_anti = any(anti in activity_text for anti in rules['anti_keywords'])

            if has_anti and ai_score < 0.75:
                continue

            has_primary = any(kw in activity_text for kw in rules['keywords'])
            if has_primary or ai_score > 0.60:
                row['AI_Match_Confidence'] = ai_score * 100
                row['Intent_Category'] = intent_category
                filtered_matches.append(row)
        else:
            row['AI_Match_Confidence'] = ai_score * 100
            row['Intent_Category'] = 'general'
            filtered_matches.append(row)

    if not filtered_matches:
        return 0, 0.0, pd.DataFrame(), pd.DataFrame()

    competitors = pd.DataFrame(filtered_matches)

    # Geospatial
    competitors['latitude'] = pd.to_numeric(competitors['latitude'], errors='coerce')
    competitors['longitude'] = pd.to_numeric(competitors['longitude'], errors='coerce')
    competitors = competitors.dropna(subset=['latitude', 'longitude'])

    if competitors.empty:
        return 0, 0.0, competitors, competitors

    competitors['Distance_km'] = haversine(target_lat, target_lon, competitors['latitude'], competitors['longitude'])
    nearby = competitors[competitors['Distance_km'] <= radius_km]

    density = len(nearby) / (math.pi * radius_km**2) if radius_km > 0 else 0

    return len(nearby), density, nearby, competitors

# ==========================================
# 5. UI (Instant Response Now!)
# ==========================================
title = widgets.HTML("<h2>⚡ Lightning-Fast MSME Analysis</h2>")
subtitle = widgets.HTML("<i>Vectors pre-computed. Search executes in milliseconds.</i>")

activity_input = widgets.Text(value='Footwear Shop', description='Query:')
lat_input = widgets.FloatText(value=9.4450, description='Lat:')
lon_input = widgets.FloatText(value=76.5400, description='Lon:')
radius_input = widgets.FloatSlider(value=2.0, min=0.5, max=10.0, step=0.5, description='Radius km:')

analyze_btn = widgets.Button(description="⚡ Analyze Now", button_style='primary')
output_area = widgets.Output()

def on_analyze(b):
    with output_area:
        clear_output()
        if df.empty:
            print("❌ No data loaded")
            return

        act = activity_input.value
        lat, lon, r = lat_input.value, lon_input.value, radius_input.value

        intent, _ = classify_business_intent(act)

        print(f"🔍 Query: '{act}' | Detected: {intent.upper()}")
        print(f"⚡ Executing vector search...")

        # ⏱️ This now takes ~0.05 seconds instead of 15 seconds!
        n_nearby, density, nearby_df, all_df = calculate_competitor_density(lat, lon, act, r)

        print(f"\n📊 RESULTS:")
        print(f"  • {n_nearby} competitors within {r}km")
        print(f"  • {len(all_df)} total matches in database")
        print(f"  • Density: {density:.2f} biz/sq.km")

        if n_nearby > 0:
            display_cols = ['EnterpriseName', 'Activities', 'AI_Match_Confidence', 'Distance_km']
            available = [c for c in display_cols if c in nearby_df.columns]
            result_df = nearby_df[available].sort_values('Distance_km')
            result_df['AI_Match_Confidence'] = result_df['AI_Match_Confidence'].round(1).astype(str) + '%'
            result_df['Distance_km'] = result_df['Distance_km'].round(3).astype(str) + ' km'
            display(result_df)

analyze_btn.on_click(on_analyze)

display(title, subtitle, activity_input, lat_input, lon_input, radius_input, analyze_btn, output_area)
print(f"\n{'='*50}")
print(f"🎯 SYSTEM READY | {len(df)} records with pre-computed vectors")
print(f"{'='*50}")
```

```python
import pandas as pd
import numpy as np
import math
import json
import ipywidgets as widgets
from IPython.display import display, clear_output
from sentence_transformers import SentenceTransformer, util
import torch

# ==========================================
# 1. CONFIGURATION
# ==========================================
FILE_NAME = '/content/geoode.csv'
VECTOR_COL = 'Activity_Vector'
MODEL_NAME = 'all-mpnet-base-v2'

print(f"🚀 MSME Intelligence Engine")
print(f"📊 Data: {FILE_NAME}")
print(f"🤖 Model: {MODEL_NAME}")

# ==========================================
# 2. LOAD MODEL & DATA (VECTORS PRE-COMPUTED)
# ==========================================
print(f"\n⏳ Loading AI Model...")
ai_model = SentenceTransformer(MODEL_NAME)
print(f"✅ AI Ready")

# Load data with pre-computed vectors
df = pd.read_csv(FILE_NAME)
df.columns = df.columns.str.strip()

# Standardize columns
rename_map = {}
for col in df.columns:
    if col.lower() in ['latitude', 'lat']: rename_map[col] = 'latitude'
    elif col.lower() in ['longitude', 'lon', 'lng']: rename_map[col] = 'longitude'
    elif col.lower() in ['activities', 'activity', 'sector']: rename_map[col] = 'Activities'
    elif col.lower() in ['enterprisename', 'name', 'enterprise_name']: rename_map[col] = 'EnterpriseName'
    elif col.lower() in ['communicationaddress', 'address']: rename_map[col] = 'CommunicationAddress'

df.rename(columns=rename_map, inplace=True)
df['Activities'] = df['Activities'].fillna("Unknown").astype(str)

# Load pre-computed vectors
print(f"💾 Loading pre-computed vectors...")
vectors = []
for vec_str in df[VECTOR_COL]:
    try:
        vectors.append(np.array(json.loads(vec_str), dtype=np.float32))
    except:
        vectors.append(np.zeros(768))  # Fallback for mpnet-base

activity_embeddings = torch.tensor(np.array(vectors), dtype=torch.float32)
print(f"✅ Loaded {len(df)} records instantly!")

# ==========================================
# 3. BUSINESS LOGIC ENGINE
# ==========================================
BUSINESS_CATEGORIES = {
    'retail': {
        'keywords': ['shop', 'store', 'retail', 'dealer', 'seller', 'sales', 'boutique', 'mart', 'showroom', 'reseller'],
        'anti_keywords': ['repair', 'service', 'mending', 'manufacturing', 'mfg', 'factory', 'industry', 'works', 'fabrication', 'production']
    },
    'service': {
        'keywords': ['repair', 'service', 'fixing', 'mending', 'alteration', 'maintenance', 'servicing', 'cobbler'],
        'anti_keywords': ['manufacturing', 'factory', 'mfg', 'production', 'industry', 'mill', 'retail', 'store', 'seller']
    },
    'manufacturing': {
        'keywords': ['manufacturing', 'mfg', 'factory', 'industry', 'production', 'mill', 'works', 'fabrication', 'maker', 'unit'],
        'anti_keywords': ['repair', 'service', 'retail', 'shop', 'store', 'seller', 'dealer']
    }
}

def classify_business_intent(query):
    query_lower = query.lower()
    scores = {}

    for category, data in BUSINESS_CATEGORIES.items():
        score = sum(2 for keyword in data['keywords'] if keyword in query_lower)
        penalty = sum(1 for anti in data['anti_keywords'] if anti in query_lower)
        final_score = score - penalty
        if final_score > 0:
            scores[category] = final_score

    if not scores:
        return 'general', 0.0

    return max(scores, key=scores.get), scores[max(scores, key=scores.get)]

def haversine(lat1, lon1, lat2, lon2):
    R = 6371.0
    phi1, phi2 = np.radians(lat1), np.radians(lat2)
    d_phi = np.radians(lat2 - lat1)
    d_lambda = np.radians(lon2 - lon1)
    a = np.sin(d_phi/2)**2 + np.cos(phi1) * np.cos(phi2) * np.sin(d_lambda/2)**2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1-a))
    return R * c

def search_competitors(target_lat, target_lon, search_query, radius_km=2.0, ai_threshold=0.45):
    """
    Returns: (nearby_competitors_df, all_matches_df, density)
    """
    if df.empty:
        return pd.DataFrame(), pd.DataFrame(), 0.0

    intent_category, _ = classify_business_intent(search_query)

    # Encode only the query (fast!)
    query_embedding = ai_model.encode(search_query, convert_to_tensor=True)

    # Search pre-computed vectors
    cosine_scores = util.cos_sim(query_embedding, activity_embeddings)[0]

    # Get all matches above threshold
    all_matching_indices = np.where(cosine_scores.cpu().numpy() >= ai_threshold)[0]

    if len(all_matching_indices) == 0:
        return pd.DataFrame(), pd.DataFrame(), 0.0

    # Build all matches dataframe
    all_matches = df.iloc[all_matching_indices].copy()
    all_matches['AI_Score'] = cosine_scores[all_matching_indices].cpu().numpy()
    all_matches['AI_Confidence'] = (all_matches['AI_Score'] * 100).round(1)

    # Apply business logic filtering
    filtered_indices = []

    for idx, row in all_matches.iterrows():
        activity_text = str(row['Activities']).lower()
        ai_score = row['AI_Score']

        if intent_category != 'general':
            rules = BUSINESS_CATEGORIES[intent_category]
            has_anti = any(anti in activity_text for anti in rules['anti_keywords'])

            # Skip if has anti-keyword unless very high confidence
            if has_anti and ai_score < 0.75:
                continue

            has_primary = any(kw in activity_text for kw in rules['keywords'])
            if has_primary or ai_score > 0.60:
                filtered_indices.append(idx)
        else:
            filtered_indices.append(idx)

    if not filtered_indices:
        return pd.DataFrame(), pd.DataFrame(), 0.0

    filtered_matches = all_matches.loc[filtered_indices].copy()

    # Geospatial filtering for nearby competitors
    filtered_matches['latitude'] = pd.to_numeric(filtered_matches['latitude'], errors='coerce')
    filtered_matches['longitude'] = pd.to_numeric(filtered_matches['longitude'], errors='coerce')
    filtered_matches = filtered_matches.dropna(subset=['latitude', 'longitude'])

    if filtered_matches.empty:
        return pd.DataFrame(), all_matches, 0.0

    # Calculate distances
    filtered_matches['Distance_km'] = haversine(
        target_lat, target_lon,
        filtered_matches['latitude'],
        filtered_matches['longitude']
    )

    # Split into nearby vs all
    nearby = filtered_matches[filtered_matches['Distance_km'] <= radius_km].copy()

    # Calculate density
    area = math.pi * (radius_km ** 2)
    density = len(nearby) / area if area > 0 else 0

    return nearby, filtered_matches, density

# ==========================================
# 4. UI - FIXED SLIDER & DISPLAY
# ==========================================
title = widgets.HTML("<h2>⚡ MSME Market Intelligence</h2>")

activity_input = widgets.Text(value='Footwear Manufacturer', description='Search:', layout=widgets.Layout(width='400px'))
lat_input = widgets.FloatText(value=9.4450, description='Latitude:', step=0.0001)
lon_input = widgets.FloatText(value=76.5400, description='Longitude:', step=0.0001)

# FIXED: Finer step control for radius
radius_input = widgets.FloatSlider(
    value=0.5,
    min=0.1,
    max=10.0,
    step=0.1,  # Now supports 0.1, 0.2, 0.5 etc
    description='Radius km:',
    readout_format='.1f'
)

analyze_btn = widgets.Button(description="⚡ Analyze", button_style='success')
output_area = widgets.Output()

def on_analyze(b):
    with output_area:
        clear_output()

        query = activity_input.value
        lat, lon, radius = lat_input.value, lon_input.value, radius_input.value

        intent, _ = classify_business_intent(query)

        print(f"{'='*60}")
        print(f"🔍 SEARCH: '{query}'")
        print(f"📍 LOCATION: ({lat}, {lon}) | RADIUS: {radius}km")
        print(f"🎯 DETECTED INTENT: {intent.upper()}")
        print(f"{'='*60}\n")

        # Execute search
        nearby_df, all_matches_df, density = search_competitors(lat, lon, query, radius)

        # SUMMARY STATS
        print(f"📊 SUMMARY:")
        print(f"   • Competitors within {radius}km: {len(nearby_df)}")
        print(f"   • Total matches in database: {len(all_matches_df)}")
        print(f"   • Market Density: {density:.2f} businesses/km²")
        print()

        # DISPLAY NEARBY COMPETITORS
        print(f"{'='*60}")
        print(f"🗺️  COMPETITORS WITHIN {radius}km RADIUS")
        print(f"{'='*60}")

        if len(nearby_df) > 0:
            display_cols = ['EnterpriseName', 'Activities', 'AI_Confidence', 'Distance_km']
            available_cols = [c for c in display_cols if c in nearby_df.columns]

            display_df = nearby_df[available_cols].sort_values('Distance_km')
            display_df = display_df.copy()
            display_df['Distance_km'] = display_df['Distance_km'].round(3).astype(str) + ' km'
            display_df['AI_Confidence'] = display_df['AI_Confidence'].astype(str) + '%'

            display(display_df)
        else:
            print("   ❌ No competitors found in this radius")

        # DISPLAY ALL DATABASE MATCHES
        print(f"\n{'='*60}")
        print(f"🌐 ALL MATCHES IN DATABASE (Regardless of Distance)")
        print(f"{'='*60}")

        if len(all_matches_df) > 0:
            display_cols = ['EnterpriseName', 'Activities', 'AI_Confidence', 'Distance_km']
            available_cols = [c for c in display_cols if c in all_matches_df.columns]

            all_display = all_matches_df[available_cols].sort_values(['AI_Confidence', 'Distance_km'], ascending=[False, True])
            all_display = all_display.copy()
            all_display['Distance_km'] = all_display['Distance_km'].round(3).astype(str) + ' km'
            all_display['AI_Confidence'] = all_display['AI_Confidence'].astype(str) + '%'

            display(all_display)
        else:
            print("   ❌ No matches found in database")

analyze_btn.on_click(on_analyze)

# Layout
inputs = widgets.VBox([activity_input, lat_input, lon_input, radius_input, analyze_btn])
display(title, inputs, output_area)

print(f"\n{'='*60}")
print(f"✅ SYSTEM READY | {len(df)} records loaded")
print(f"⚡ Vector search active | Pre-computed embeddings")
print(f"{'='*60}")
```

```python
import pandas as pd
import numpy as np
import math
import json
import ipywidgets as widgets
from IPython.display import display, clear_output
from sentence_transformers import SentenceTransformer, util
import torch

# ==========================================
# 1. CONFIGURATION
# ==========================================
FILE_NAME = '/content/geoode.csv'
VECTOR_COL = 'Activity_Vector'
MODEL_NAME = 'all-mpnet-base-v2'

print(f"🚀 MSME Intelligence Engine")
print(f"📊 Data: {FILE_NAME}")
print(f"🤖 Model: {MODEL_NAME}")

# ==========================================
# 2. LOAD MODEL & DATA (VECTORS PRE-COMPUTED)
# ==========================================
print(f"\n⏳ Loading AI Model...")
ai_model = SentenceTransformer(MODEL_NAME)
print(f"✅ AI Ready")

# Load data with pre-computed vectors
df = pd.read_csv(FILE_NAME)
df.columns = df.columns.str.strip()

# Standardize columns
rename_map = {}
for col in df.columns:
    if col.lower() in ['latitude', 'lat']: rename_map[col] = 'latitude'
    elif col.lower() in ['longitude', 'lon', 'lng']: rename_map[col] = 'longitude'
    elif col.lower() in ['activities', 'activity', 'sector']: rename_map[col] = 'Activities'
    elif col.lower() in ['enterprisename', 'name', 'enterprise_name']: rename_map[col] = 'EnterpriseName'
    elif col.lower() in ['communicationaddress', 'address']: rename_map[col] = 'CommunicationAddress'

df.rename(columns=rename_map, inplace=True)
df['Activities'] = df['Activities'].fillna("Unknown").astype(str)

# Ensure coordinates are numeric for new density/distance calculations
if 'latitude' in df.columns and 'longitude' in df.columns:
    df['latitude'] = pd.to_numeric(df['latitude'], errors='coerce')
    df['longitude'] = pd.to_numeric(df['longitude'], errors='coerce')

# Load pre-computed vectors
print(f"💾 Loading pre-computed vectors...")
vectors = []
for vec_str in df[VECTOR_COL]:
    try:
        vectors.append(np.array(json.loads(vec_str), dtype=np.float32))
    except:
        vectors.append(np.zeros(768))  # Fallback for mpnet-base

activity_embeddings = torch.tensor(np.array(vectors), dtype=torch.float32)
print(f"✅ Loaded {len(df)} records instantly!")

# ==========================================
# 3. BUSINESS LOGIC ENGINE
# ==========================================
BUSINESS_CATEGORIES = {
    'retail': {
        'keywords': ['shop', 'store', 'retail', 'dealer', 'seller', 'sales', 'boutique', 'mart', 'showroom', 'reseller'],
        'anti_keywords': ['repair', 'service', 'mending', 'manufacturing', 'mfg', 'factory', 'industry', 'works', 'fabrication', 'production']
    },
    'service': {
        'keywords': ['repair', 'service', 'fixing', 'mending', 'alteration', 'maintenance', 'servicing', 'cobbler'],
        'anti_keywords': ['manufacturing', 'factory', 'mfg', 'production', 'industry', 'mill', 'retail', 'store', 'seller']
    },
    'manufacturing': {
        'keywords': ['manufacturing', 'mfg', 'factory', 'industry', 'production', 'mill', 'works', 'fabrication', 'maker', 'unit'],
        'anti_keywords': ['repair', 'service', 'retail', 'shop', 'store', 'seller', 'dealer']
    }
}

# --- NEW: MAGNETIC POLES FOR PINCODE 686101 (Changanassery) ---
MAGNETIC_POLES = {
    "Changanassery Railway Station": (9.4522, 76.5451),
    "Changanassery KSRTC Bus Stand": (9.4447, 76.5385)
}

def classify_business_intent(query):
    query_lower = query.lower()
    scores = {}

    for category, data in BUSINESS_CATEGORIES.items():
        score = sum(2 for keyword in data['keywords'] if keyword in query_lower)
        penalty = sum(1 for anti in data['anti_keywords'] if anti in query_lower)
        final_score = score - penalty
        if final_score > 0:
            scores[category] = final_score

    if not scores:
        return 'general', 0.0

    return max(scores, key=scores.get), scores[max(scores, key=scores.get)]

def haversine(lat1, lon1, lat2, lon2):
    R = 6371.0
    phi1, phi2 = np.radians(lat1), np.radians(lat2)
    d_phi = np.radians(lat2 - lat1)
    d_lambda = np.radians(lon2 - lon1)
    a = np.sin(d_phi/2)**2 + np.cos(phi1) * np.cos(phi2) * np.sin(d_lambda/2)**2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1-a))
    return R * c

# --- NEW: HOTSPOT DETECTION (DBSCAN LOGIC) ---
def check_hotspot_status(target_lat, target_lon, epsilon=0.5, min_pts=3):
    """Checks if the target location is a core point (hotspot) based on ALL businesses."""
    if df.empty or 'latitude' not in df.columns or 'longitude' not in df.columns:
        return False, 0

    # Calculate distance to all known businesses in dataset
    distances = haversine(
        target_lat, target_lon,
        df['latitude'],
        df['longitude']
    )

    # Neighborhood size |N_eps(p)|
    neighborhood_size = (distances <= epsilon).sum()

    # DBSCAN Core Point Condition
    is_hotspot = neighborhood_size >= min_pts
    return is_hotspot, neighborhood_size

def search_competitors(target_lat, target_lon, search_query, radius_km=2.0, ai_threshold=0.45):
    """
    Returns: (nearby_competitors_df, all_matches_df, density)
    """
    if df.empty:
        return pd.DataFrame(), pd.DataFrame(), 0.0

    intent_category, _ = classify_business_intent(search_query)

    # Encode only the query (fast!)
    query_embedding = ai_model.encode(search_query, convert_to_tensor=True)

    # Search pre-computed vectors
    cosine_scores = util.cos_sim(query_embedding, activity_embeddings)[0]

    # Get all matches above threshold
    all_matching_indices = np.where(cosine_scores.cpu().numpy() >= ai_threshold)[0]

    if len(all_matching_indices) == 0:
        return pd.DataFrame(), pd.DataFrame(), 0.0

    # Build all matches dataframe
    all_matches = df.iloc[all_matching_indices].copy()
    all_matches['AI_Score'] = cosine_scores[all_matching_indices].cpu().numpy()
    all_matches['AI_Confidence'] = (all_matches['AI_Score'] * 100).round(1)

    # Apply business logic filtering
    filtered_indices = []

    for idx, row in all_matches.iterrows():
        activity_text = str(row['Activities']).lower()
        ai_score = row['AI_Score']

        if intent_category != 'general':
            rules = BUSINESS_CATEGORIES[intent_category]
            has_anti = any(anti in activity_text for anti in rules['anti_keywords'])

            # Skip if has anti-keyword unless very high confidence
            if has_anti and ai_score < 0.75:
                continue

            has_primary = any(kw in activity_text for kw in rules['keywords'])
            if has_primary or ai_score > 0.60:
                filtered_indices.append(idx)
        else:
            filtered_indices.append(idx)

    if not filtered_indices:
        return pd.DataFrame(), pd.DataFrame(), 0.0

    filtered_matches = all_matches.loc[filtered_indices].copy()

    # Geospatial filtering for nearby competitors
    filtered_matches = filtered_matches.dropna(subset=['latitude', 'longitude'])

    if filtered_matches.empty:
        return pd.DataFrame(), all_matches, 0.0

    # Calculate distances
    filtered_matches['Distance_km'] = haversine(
        target_lat, target_lon,
        filtered_matches['latitude'],
        filtered_matches['longitude']
    )

    # Split into nearby vs all
    nearby = filtered_matches[filtered_matches['Distance_km'] <= radius_km].copy()

    # Calculate density
    area = math.pi * (radius_km ** 2)
    density = len(nearby) / area if area > 0 else 0

    return nearby, filtered_matches, density

# ==========================================
# 4. UI - FIXED SLIDER & DISPLAY
# ==========================================
title = widgets.HTML("<h2>⚡ MSME Market Intelligence</h2>")

activity_input = widgets.Text(value='Footwear Manufacturer', description='Search:', layout=widgets.Layout(width='400px'))
lat_input = widgets.FloatText(value=9.4450, description='Latitude:', step=0.0001)
lon_input = widgets.FloatText(value=76.5400, description='Longitude:', step=0.0001)

# FIXED: Finer step control for radius
radius_input = widgets.FloatSlider(
    value=0.5,
    min=0.1,
    max=10.0,
    step=0.1,  # Now supports 0.1, 0.2, 0.5 etc
    description='Radius km:',
    readout_format='.1f'
)

analyze_btn = widgets.Button(description="⚡ Analyze", button_style='success')
output_area = widgets.Output()

def on_analyze(b):
    with output_area:
        clear_output()

        query = activity_input.value
        lat, lon, radius = lat_input.value, lon_input.value, radius_input.value

        intent, _ = classify_business_intent(query)

        print(f"{'='*60}")
        print(f"🔍 SEARCH: '{query}'")
        print(f"📍 LOCATION: ({lat}, {lon}) | RADIUS: {radius}km")
        print(f"🎯 DETECTED INTENT: {intent.upper()}")
        print(f"{'='*60}\n")

        # Execute competitor search
        nearby_df, all_matches_df, density = search_competitors(lat, lon, query, radius)

        # --- EXECUTE NEW FEATURES ---
        # 1. DBSCAN Hotspot Detection
        is_hotspot, total_nearby_businesses = check_hotspot_status(lat, lon, epsilon=0.5, min_pts=3)
        hotspot_label = "✅ YES (Thriving Commercial Cluster)" if is_hotspot else "⚠️ NO (Isolated Area)"

        # 2. Magnetic Poles Distances (Logistics)
        pole_distances = {}
        for pole, coords in MAGNETIC_POLES.items():
            pole_distances[pole] = haversine(lat, lon, coords[0], coords[1])

        # SUMMARY STATS (Updated with new features)
        print(f"📊 SUMMARY & SPATIAL INTELLIGENCE:")
        print(f"   • Competitors within {radius}km: {len(nearby_df)}")
        print(f"   • Total matches in database: {len(all_matches_df)}")
        print(f"   • Market Density: {density:.2f} competitors/km²")
        print(f"   • Commercial Hotspot (ε=0.5km): {hotspot_label} [{total_nearby_businesses} total businesses nearby]")
        print(f"   • Logistics & Footfall (Distance to Magnetic Poles):")
        for pole, dist in pole_distances.items():
            print(f"       - {pole}: {dist:.2f} km")
        print()

        # DISPLAY NEARBY COMPETITORS
        print(f"{'='*60}")
        print(f"🗺️  COMPETITORS WITHIN {radius}km RADIUS")
        print(f"{'='*60}")

        if len(nearby_df) > 0:
            display_cols = ['EnterpriseName', 'Activities', 'AI_Confidence', 'Distance_km']
            available_cols = [c for c in display_cols if c in nearby_df.columns]

            display_df = nearby_df[available_cols].sort_values('Distance_km')
            display_df = display_df.copy()
            display_df['Distance_km'] = display_df['Distance_km'].round(3).astype(str) + ' km'
            display_df['AI_Confidence'] = display_df['AI_Confidence'].astype(str) + '%'

            display(display_df)
        else:
            print("   ❌ No competitors found in this radius")

        # DISPLAY ALL DATABASE MATCHES
        print(f"\n{'='*60}")
        print(f"🌐 ALL MATCHES IN DATABASE (Regardless of Distance)")
        print(f"{'='*60}")

        if len(all_matches_df) > 0:
            display_cols = ['EnterpriseName', 'Activities', 'AI_Confidence', 'Distance_km']
            available_cols = [c for c in display_cols if c in all_matches_df.columns]

            all_display = all_matches_df[available_cols].sort_values(['AI_Confidence', 'Distance_km'], ascending=[False, True])
            all_display = all_display.copy()
            all_display['Distance_km'] = all_display['Distance_km'].round(3).astype(str) + ' km'
            all_display['AI_Confidence'] = all_display['AI_Confidence'].astype(str) + '%'

            display(all_display)
        else:
            print("   ❌ No matches found in database")

analyze_btn.on_click(on_analyze)

# Layout
inputs = widgets.VBox([activity_input, lat_input, lon_input, radius_input, analyze_btn])
display(title, inputs, output_area)

print(f"\n{'='*60}")
print(f"✅ SYSTEM READY | {len(df)} records loaded")
print(f"⚡ Vector search active | Pre-computed embeddings")
print(f"{'='*60}")
```

```python
import pandas as pd
import numpy as np
import math
import json
import ipywidgets as widgets
from IPython.display import display, clear_output
from sentence_transformers import SentenceTransformer, util
import torch

# ==========================================
# 1. CONFIGURATION
# ==========================================
FILE_NAME = '/content/geoode (1).csv'
VECTOR_COL = 'Activity_Vector'
MODEL_NAME = 'all-mpnet-base-v2'

print(f"🚀 MSME Intelligence Engine")
print(f"📊 Data: {FILE_NAME}")
print(f"🤖 Model: {MODEL_NAME}")

# ==========================================
# 2. LOAD MODEL & DATA (VECTORS PRE-COMPUTED)
# ==========================================
print(f"\n⏳ Loading AI Model...")
ai_model = SentenceTransformer(MODEL_NAME)
print(f"✅ AI Ready")

# Load data with pre-computed vectors
df = pd.read_csv(FILE_NAME)
df.columns = df.columns.str.strip()

# Standardize columns
rename_map = {}
for col in df.columns:
    if col.lower() in ['latitude', 'lat']: rename_map[col] = 'latitude'
    elif col.lower() in ['longitude', 'lon', 'lng']: rename_map[col] = 'longitude'
    elif col.lower() in ['activities', 'activity', 'sector']: rename_map[col] = 'Activities'
    elif col.lower() in ['enterprisename', 'name', 'enterprise_name']: rename_map[col] = 'EnterpriseName'
    elif col.lower() in ['communicationaddress', 'address']: rename_map[col] = 'CommunicationAddress'

df.rename(columns=rename_map, inplace=True)
df['Activities'] = df['Activities'].fillna("Unknown").astype(str)

# Ensure coordinates are numeric for new density/distance calculations
if 'latitude' in df.columns and 'longitude' in df.columns:
    df['latitude'] = pd.to_numeric(df['latitude'], errors='coerce')
    df['longitude'] = pd.to_numeric(df['longitude'], errors='coerce')

# Load pre-computed vectors
print(f"💾 Loading pre-computed vectors...")
vectors = []
for vec_str in df[VECTOR_COL]:
    try:
        vectors.append(np.array(json.loads(vec_str), dtype=np.float32))
    except:
        vectors.append(np.zeros(768))  # Fallback for mpnet-base

activity_embeddings = torch.tensor(np.array(vectors), dtype=torch.float32)
print(f"✅ Loaded {len(df)} records instantly!")

# ==========================================
# 3. BUSINESS LOGIC ENGINE
# ==========================================
BUSINESS_CATEGORIES = {
    'retail': {
        'keywords': ['shop', 'store', 'retail', 'dealer', 'seller', 'sales', 'boutique', 'mart', 'showroom', 'reseller'],
        'anti_keywords': ['repair', 'service', 'mending', 'manufacturing', 'mfg', 'factory', 'industry', 'works', 'fabrication', 'production']
    },
    'service': {
        'keywords': ['repair', 'service', 'fixing', 'mending', 'alteration', 'maintenance', 'servicing', 'cobbler'],
        'anti_keywords': ['manufacturing', 'factory', 'mfg', 'production', 'industry', 'mill', 'retail', 'store', 'seller']
    },
    'manufacturing': {
        'keywords': ['manufacturing', 'mfg', 'factory', 'industry', 'production', 'mill', 'works', 'fabrication', 'maker', 'unit'],
        'anti_keywords': ['repair', 'service', 'retail', 'shop', 'store', 'seller', 'dealer']
    }
}

# --- EXISTING: MAGNETIC POLES FOR PINCODE 686101 (Changanassery) ---
MAGNETIC_POLES = {
    "Changanassery Railway Station": (9.4522, 76.5451),
    "Changanassery KSRTC Bus Stand": (9.4447, 76.5385)
}

def classify_business_intent(query):
    query_lower = query.lower()
    scores = {}

    for category, data in BUSINESS_CATEGORIES.items():
        score = sum(2 for keyword in data['keywords'] if keyword in query_lower)
        penalty = sum(1 for anti in data['anti_keywords'] if anti in query_lower)
        final_score = score - penalty
        if final_score > 0:
            scores[category] = final_score

    if not scores:
        return 'general', 0.0

    return max(scores, key=scores.get), scores[max(scores, key=scores.get)]

def haversine(lat1, lon1, lat2, lon2):
    R = 6371.0
    phi1, phi2 = np.radians(lat1), np.radians(lat2)
    d_phi = np.radians(lat2 - lat1)
    d_lambda = np.radians(lon2 - lon1)
    a = np.sin(d_phi/2)**2 + np.cos(phi1) * np.cos(phi2) * np.sin(d_lambda/2)**2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1-a))
    return R * c

# --- EXISTING: HOTSPOT DETECTION (DBSCAN LOGIC) ---
def check_hotspot_status(target_lat, target_lon, epsilon=0.5, min_pts=3):
    """Checks if the target location is a core point (hotspot) based on ALL businesses."""
    if df.empty or 'latitude' not in df.columns or 'longitude' not in df.columns:
        return False, 0

    distances = haversine(target_lat, target_lon, df['latitude'], df['longitude'])
    neighborhood_size = (distances <= epsilon).sum()
    is_hotspot = neighborhood_size >= min_pts
    return is_hotspot, neighborhood_size

# --- NEW: HUFF MODEL (RETAIL GRAVITY) ---
def calculate_huff_gravity(nearby_distances, lambda_param=2.0, internal_radius=0.1):
    """
    Calculates the probability of local customers choosing the new store over competitors.
    Assumes customer is in the immediate vicinity (internal_radius).
    """
    if len(nearby_distances) == 0:
        return 1.0  # 100% market share if no competitors

    # Gravity of our new store (Size S=1, Distance = internal_radius)
    our_gravity = 1.0 / (internal_radius ** lambda_param)

    # Gravity of competitors pulling customers away (Size S=1, Distance = actual distance)
    # Use max to prevent division by zero if a competitor is located at the exact same coordinates
    safe_distances = np.maximum(nearby_distances, 0.01)
    competitor_gravity_sum = np.sum(1.0 / (safe_distances ** lambda_param))

    # Huff Probability
    probability = our_gravity / (our_gravity + competitor_gravity_sum)
    return probability

# --- NEW: WHITE SPACE ANALYSIS ---
def calculate_white_space(all_matches_df, radius_km):
    """Calculates d_min (distance to closest competitor) and Density D."""
    if all_matches_df.empty:
        return float('inf'), 0.0

    d_min = all_matches_df['Distance_km'].min()

    # Density D = N / (pi * R^2)
    N = len(all_matches_df[all_matches_df['Distance_km'] <= radius_km])
    area = math.pi * (radius_km ** 2)
    density = N / area if area > 0 else 0.0

    return d_min, density

def search_competitors(target_lat, target_lon, search_query, radius_km=2.0, ai_threshold=0.45):
    """Returns: (nearby_competitors_df, all_matches_df, density)"""
    if df.empty:
        return pd.DataFrame(), pd.DataFrame(), 0.0

    intent_category, _ = classify_business_intent(search_query)

    query_embedding = ai_model.encode(search_query, convert_to_tensor=True)
    cosine_scores = util.cos_sim(query_embedding, activity_embeddings)[0]
    all_matching_indices = np.where(cosine_scores.cpu().numpy() >= ai_threshold)[0]

    if len(all_matching_indices) == 0:
        return pd.DataFrame(), pd.DataFrame(), 0.0

    all_matches = df.iloc[all_matching_indices].copy()
    all_matches['AI_Score'] = cosine_scores[all_matching_indices].cpu().numpy()
    all_matches['AI_Confidence'] = (all_matches['AI_Score'] * 100).round(1)

    filtered_indices = []
    for idx, row in all_matches.iterrows():
        activity_text = str(row['Activities']).lower()
        ai_score = row['AI_Score']

        if intent_category != 'general':
            rules = BUSINESS_CATEGORIES[intent_category]
            has_anti = any(anti in activity_text for anti in rules['anti_keywords'])

            if has_anti and ai_score < 0.75:
                continue

            has_primary = any(kw in activity_text for kw in rules['keywords'])
            if has_primary or ai_score > 0.60:
                filtered_indices.append(idx)
        else:
            filtered_indices.append(idx)

    if not filtered_indices:
        return pd.DataFrame(), pd.DataFrame(), 0.0

    filtered_matches = all_matches.loc[filtered_indices].copy()
    filtered_matches = filtered_matches.dropna(subset=['latitude', 'longitude'])

    if filtered_matches.empty:
        return pd.DataFrame(), all_matches, 0.0

    filtered_matches['Distance_km'] = haversine(
        target_lat, target_lon,
        filtered_matches['latitude'],
        filtered_matches['longitude']
    )

    nearby = filtered_matches[filtered_matches['Distance_km'] <= radius_km].copy()

    area = math.pi * (radius_km ** 2)
    density = len(nearby) / area if area > 0 else 0

    return nearby, filtered_matches, density

# ==========================================
# 4. UI - FIXED SLIDER & DISPLAY
# ==========================================
title = widgets.HTML("<h2>⚡ MSME Market Intelligence</h2>")

activity_input = widgets.Text(value='Footwear Manufacturer', description='Search:', layout=widgets.Layout(width='400px'))
lat_input = widgets.FloatText(value=9.4450, description='Latitude:', step=0.0001)
lon_input = widgets.FloatText(value=76.5400, description='Longitude:', step=0.0001)

radius_input = widgets.FloatSlider(
    value=0.5,
    min=0.1,
    max=10.0,
    step=0.1,
    description='Radius km:',
    readout_format='.1f'
)

analyze_btn = widgets.Button(description="⚡ Analyze", button_style='success')
output_area = widgets.Output()

def on_analyze(b):
    with output_area:
        clear_output()

        query = activity_input.value
        lat, lon, radius = lat_input.value, lon_input.value, radius_input.value

        intent, _ = classify_business_intent(query)

        print(f"{'='*60}")
        print(f"🔍 SEARCH: '{query}'")
        print(f"📍 LOCATION: ({lat}, {lon}) | RADIUS: {radius}km")
        print(f"🎯 DETECTED INTENT: {intent.upper()}")
        print(f"{'='*60}\n")

        # Execute competitor search
        nearby_df, all_matches_df, old_density = search_competitors(lat, lon, query, radius)

        # --- EXECUTE GEOSPATIAL FEATURES ---
        # 1. DBSCAN Hotspot Detection
        is_hotspot, total_nearby_businesses = check_hotspot_status(lat, lon, epsilon=0.5, min_pts=3)
        hotspot_label = "✅ YES (Thriving Commercial Cluster)" if is_hotspot else "⚠️ NO (Isolated Area)"

        # 2. Magnetic Poles Distances (Logistics)
        pole_distances = {}
        for pole, coords in MAGNETIC_POLES.items():
            pole_distances[pole] = haversine(lat, lon, coords[0], coords[1])

        # 3. White Space Analysis
        d_min, new_density = calculate_white_space(all_matches_df, radius)
        if d_min == float('inf'):
            d_min_str = "No competitors globally"
            ws_status = "✅ Pristine White Space (Absolute Monopoly)"
        else:
            d_min_str = f"{d_min:.3f} km"
            # If the closest competitor is further than half our radius, it's a good white space
            ws_status = "✅ Strong (High Monopoly Potential)" if d_min >= (radius * 0.5) else "⚠️ Weak (Highly Competitive Area)"

        # 4. Huff Model (Retail Gravity)
        if len(nearby_df) > 0:
            huff_prob = calculate_huff_gravity(nearby_df['Distance_km'].values)
        else:
            huff_prob = 1.0

        # SUMMARY STATS (Updated with all features)
        print(f"📊 SUMMARY & SPATIAL INTELLIGENCE:")
        print(f"   • Competitors within {radius}km: {len(nearby_df)}")
        print(f"   • Market Density (D): {new_density:.2f} competitors/km²")
        print(f"   • Commercial Hotspot (ε=0.5km): {hotspot_label} [{total_nearby_businesses} businesses nearby]")
        print()
        print(f"   • White Space Analysis:")
        print(f"       - Nearest Direct Competitor (d_min): {d_min_str}")
        print(f"       - Status: {ws_status}")
        print()
        print(f"   • Retail Gravity (Huff Model):")
        print(f"       - Est. Immediate Market Capture: {huff_prob * 100:.1f}%")
        print()
        print(f"   • Logistics & Footfall (Dist. to Magnetic Poles):")
        for pole, dist in pole_distances.items():
            print(f"       - {pole}: {dist:.2f} km")
        print()

        # DISPLAY NEARBY COMPETITORS
        print(f"{'='*60}")
        print(f"🗺️  COMPETITORS WITHIN {radius}km RADIUS")
        print(f"{'='*60}")

        if len(nearby_df) > 0:
            display_cols = ['EnterpriseName', 'Activities', 'AI_Confidence', 'Distance_km']
            available_cols = [c for c in display_cols if c in nearby_df.columns]

            display_df = nearby_df[available_cols].sort_values('Distance_km')
            display_df = display_df.copy()
            display_df['Distance_km'] = display_df['Distance_km'].round(3).astype(str) + ' km'
            display_df['AI_Confidence'] = display_df['AI_Confidence'].astype(str) + '%'

            display(display_df)
        else:
            print("   ❌ No competitors found in this radius")

        # DISPLAY ALL DATABASE MATCHES
        print(f"\n{'='*60}")
        print(f"🌐 ALL MATCHES IN DATABASE (Regardless of Distance)")
        print(f"{'='*60}")

        if len(all_matches_df) > 0:
            display_cols = ['EnterpriseName', 'Activities', 'AI_Confidence', 'Distance_km']
            available_cols = [c for c in display_cols if c in all_matches_df.columns]

            all_display = all_matches_df[available_cols].sort_values(['AI_Confidence', 'Distance_km'], ascending=[False, True])
            all_display = all_display.copy()
            all_display['Distance_km'] = all_display['Distance_km'].round(3).astype(str) + ' km'
            all_display['AI_Confidence'] = all_display['AI_Confidence'].astype(str) + '%'

            display(all_display)
        else:
            print("   ❌ No matches found in database")

analyze_btn.on_click(on_analyze)

# Layout
inputs = widgets.VBox([activity_input, lat_input, lon_input, radius_input, analyze_btn])
display(title, inputs, output_area)

print(f"\n{'='*60}")
print(f"✅ SYSTEM READY | {len(df)} records loaded")
print(f"⚡ Vector search active | Pre-computed embeddings")
print(f"{'='*60}")
```

```python
%%writefile app.py
import streamlit as st
import pandas as pd
import numpy as np
import math
import json
from sentence_transformers import SentenceTransformer, util
import torch
import plotly.express as px

st.set_page_config(page_title="MSME Intelligence Engine", layout="wide", page_icon="🏦")

# ==========================================
# 1. CACHED DATA & MODEL LOADING
# ==========================================
@st.cache_resource
def load_system():
    FILE_NAME = '/content/geoode (1).csv'
    MODEL_NAME = 'all-mpnet-base-v2'

    ai_model = SentenceTransformer(MODEL_NAME)
    df = pd.read_csv(FILE_NAME)
    df.columns = df.columns.str.strip()

    rename_map = {}
    for col in df.columns:
        if col.lower() in ['latitude', 'lat']: rename_map[col] = 'latitude'
        elif col.lower() in ['longitude', 'lon', 'lng']: rename_map[col] = 'longitude'
        elif col.lower() in ['activities', 'activity', 'sector']: rename_map[col] = 'Activities'
        elif col.lower() in ['enterprisename', 'name', 'enterprise_name']: rename_map[col] = 'EnterpriseName'
        elif col.lower() in ['communicationaddress', 'address']: rename_map[col] = 'CommunicationAddress'

    df.rename(columns=rename_map, inplace=True)
    df['Activities'] = df['Activities'].fillna("Unknown").astype(str)

    if 'latitude' in df.columns and 'longitude' in df.columns:
        df['latitude'] = pd.to_numeric(df['latitude'], errors='coerce')
        df['longitude'] = pd.to_numeric(df['longitude'], errors='coerce')

    vectors = []
    for vec_str in df.get('Activity_Vector', []):
        try:
            vectors.append(np.array(json.loads(vec_str), dtype=np.float32))
        except:
            vectors.append(np.zeros(768))

    activity_embeddings = torch.tensor(np.array(vectors), dtype=torch.float32)
    return ai_model, df, activity_embeddings

ai_model, df, activity_embeddings = load_system()

# ==========================================
# 2. BUSINESS LOGIC ENGINE
# ==========================================
BUSINESS_CATEGORIES = {
    'retail': {'keywords': ['shop', 'store', 'retail', 'dealer', 'seller', 'sales', 'boutique', 'mart', 'showroom', 'reseller'], 'anti_keywords': ['repair', 'service', 'mending', 'manufacturing', 'mfg', 'factory', 'industry', 'works', 'fabrication', 'production']},
    'service': {'keywords': ['repair', 'service', 'fixing', 'mending', 'alteration', 'maintenance', 'servicing', 'cobbler'], 'anti_keywords': ['manufacturing', 'factory', 'mfg', 'production', 'industry', 'mill', 'retail', 'store', 'seller']},
    'manufacturing': {'keywords': ['manufacturing', 'mfg', 'factory', 'industry', 'production', 'mill', 'works', 'fabrication', 'maker', 'unit'], 'anti_keywords': ['repair', 'service', 'retail', 'shop', 'store', 'seller', 'dealer']}
}

MAGNETIC_POLES = {"Changanassery Railway Station": (9.4522, 76.5451), "Changanassery KSRTC Bus Stand": (9.4447, 76.5385)}

def classify_business_intent(query):
    query_lower = query.lower()
    scores = {}
    for category, data in BUSINESS_CATEGORIES.items():
        score = sum(2 for keyword in data['keywords'] if keyword in query_lower)
        penalty = sum(1 for anti in data['anti_keywords'] if anti in query_lower)
        final_score = score - penalty
        if final_score > 0: scores[category] = final_score
    if not scores: return 'general', 0.0
    return max(scores, key=scores.get), scores[max(scores, key=scores.get)]

def haversine(lat1, lon1, lat2, lon2):
    R = 6371.0
    phi1, phi2 = np.radians(lat1), np.radians(lat2)
    d_phi = np.radians(lat2 - lat1)
    d_lambda = np.radians(lon2 - lon1)
    a = np.sin(d_phi/2)**2 + np.cos(phi1) * np.cos(phi2) * np.sin(d_lambda/2)**2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1-a))
    return R * c

def check_hotspot_status(target_lat, target_lon, epsilon=0.5, min_pts=3):
    if df.empty or 'latitude' not in df.columns or 'longitude' not in df.columns: return False, 0
    distances = haversine(target_lat, target_lon, df['latitude'], df['longitude'])
    neighborhood_size = (distances <= epsilon).sum()
    return neighborhood_size >= min_pts, neighborhood_size

def calculate_huff_gravity(nearby_distances, lambda_param=2.0, internal_radius=0.1):
    if len(nearby_distances) == 0: return 1.0
    our_gravity = 1.0 / (internal_radius ** lambda_param)
    safe_distances = np.maximum(nearby_distances, 0.01)
    competitor_gravity_sum = np.sum(1.0 / (safe_distances ** lambda_param))
    return our_gravity / (our_gravity + competitor_gravity_sum)

def calculate_white_space(all_matches_df, radius_km):
    if all_matches_df.empty: return float('inf'), 0.0
    d_min = all_matches_df['Distance_km'].min()
    N = len(all_matches_df[all_matches_df['Distance_km'] <= radius_km])
    area = math.pi * (radius_km ** 2)
    return d_min, N / area if area > 0 else 0.0

def search_competitors(target_lat, target_lon, search_query, radius_km=2.0, ai_threshold=0.45):
    if df.empty: return pd.DataFrame(), pd.DataFrame()
    intent_category, _ = classify_business_intent(search_query)

    query_embedding = ai_model.encode(search_query, convert_to_tensor=True)
    cosine_scores = util.cos_sim(query_embedding, activity_embeddings)[0]
    all_matching_indices = np.where(cosine_scores.cpu().numpy() >= ai_threshold)[0]

    if len(all_matching_indices) == 0: return pd.DataFrame(), pd.DataFrame()

    all_matches = df.iloc[all_matching_indices].copy()
    all_matches['AI_Score'] = cosine_scores[all_matching_indices].cpu().numpy()
    all_matches['AI_Confidence'] = (all_matches['AI_Score'] * 100).round(1)

    filtered_indices = []
    for idx, row in all_matches.iterrows():
        activity_text = str(row['Activities']).lower()
        ai_score = row['AI_Score']
        if intent_category != 'general':
            rules = BUSINESS_CATEGORIES[intent_category]
            has_anti = any(anti in activity_text for anti in rules['anti_keywords'])
            if has_anti and ai_score < 0.75: continue
            has_primary = any(kw in activity_text for kw in rules['keywords'])
            if has_primary or ai_score > 0.60: filtered_indices.append(idx)
        else:
            filtered_indices.append(idx)

    if not filtered_indices: return pd.DataFrame(), pd.DataFrame()
    filtered_matches = all_matches.loc[filtered_indices].copy()
    filtered_matches = filtered_matches.dropna(subset=['latitude', 'longitude'])

    if filtered_matches.empty: return pd.DataFrame(), filtered_matches
    filtered_matches['Distance_km'] = haversine(target_lat, target_lon, filtered_matches['latitude'], filtered_matches['longitude'])

    nearby = filtered_matches[filtered_matches['Distance_km'] <= radius_km].copy()
    return nearby, filtered_matches

# --- NEW: LOAN ELIGIBILITY SCORING ENGINE ---
def calculate_loan_eligibility(huff_prob, is_hotspot, density, pole_distances):
    score = 0

    # 1. Market Capture (Max 35 points) - High Huff prob is great
    score += (huff_prob * 35)

    # 2. Hotspot / Footfall (Max 15 points)
    if is_hotspot: score += 15

    # 3. Market Saturation (Max 30 points) - Lower density is better
    density_score = max(0, 30 - (density * 5))
    score += density_score

    # 4. Logistics (Max 20 points) - Closer to magnetic poles is better
    avg_pole_dist = sum(pole_distances.values()) / len(pole_distances) if pole_distances else 5.0
    logistics_score = max(0, 20 - (avg_pole_dist * 2.5))
    score += logistics_score

    # Bound score between 0 and 100
    final_score = min(100, max(0, score))

    if final_score >= 75:
        verdict = "✅ HIGH VIABILITY (Recommend Approval)"
        color = "green"
        reasoning = "Excellent spatial positioning. Low competitive friction and strong market capture potential."
    elif final_score >= 50:
        verdict = "⚠️ MODERATE VIABILITY (Requires Scrutiny)"
        color = "orange"
        reasoning = "Acceptable location, but faces competitive pressure. Ensure applicant has strong differentiation or financial collateral."
    else:
        verdict = "❌ HIGH RISK (Geospatially Unviable)"
        color = "red"
        reasoning = "Market is heavily saturated or location is highly isolated. Severe risk of business failure due to geographic factors."

    return int(final_score), verdict, reasoning, color

# ==========================================
# 3. STREAMLIT UI & DASHBOARD
# ==========================================
st.title("🏦 Bank MSME Intelligence & Underwriting Dashboard")
st.markdown("Geospatial Credit Risk Engine for **Pincode 686101**")

with st.sidebar:
    st.header("📍 Application Parameters")
    query = st.text_input("Proposed Business Activity:", value='Footwear Manufacturer')
    lat = st.number_input("Proposed Latitude:", value=9.4450, format="%.4f")
    lon = st.number_input("Proposed Longitude:", value=76.5400, format="%.4f")
    radius = st.slider("Trade Area Radius (km):", min_value=0.1, max_value=10.0, value=2.0, step=0.1)
    analyze_btn = st.button("⚡ Run Loan Risk Assessment", type="primary", use_container_width=True)

if analyze_btn:
    # Run all backend calculations
    intent, _ = classify_business_intent(query)
    nearby_df, all_matches_df = search_competitors(lat, lon, query, radius)
    is_hotspot, total_nearby_businesses = check_hotspot_status(lat, lon, epsilon=0.5, min_pts=3)
    d_min, market_density = calculate_white_space(all_matches_df, radius)
    huff_prob = calculate_huff_gravity(nearby_df['Distance_km'].values) if len(nearby_df) > 0 else 1.0

    pole_distances = {pole: haversine(lat, lon, coords[0], coords[1]) for pole, coords in MAGNETIC_POLES.items()}

    # Calculate Bank Eligibility Score
    spatial_score, verdict, reasoning, color = calculate_loan_eligibility(huff_prob, is_hotspot, market_density, pole_distances)

    # Create Layout Tabs
    tab1, tab2, tab3 = st.tabs(["📑 Bank Decision Engine", "📈 Deep Analytics & Visuals", "🗄️ Raw Data"])

    # --- TAB 1: BANK DECISION ENGINE (EXECUTIVE SUMMARY) ---
    with tab1:
        # --- NEW BANK UI COMPONENT ---
        st.markdown(f"### 🎯 Geospatial Loan Viability Score")
        st.progress(spatial_score / 100)

        if color == "green": st.success(f"**Score: {spatial_score}/100** — {verdict}\n\n*Underwriter Note:* {reasoning}")
        elif color == "orange": st.warning(f"**Score: {spatial_score}/100** — {verdict}\n\n*Underwriter Note:* {reasoning}")
        else: st.error(f"**Score: {spatial_score}/100** — {verdict}\n\n*Underwriter Note:* {reasoning}")

        st.divider()

        st.subheader(f"Geospatial Metrics: {query.title()}")
        col1, col2, col3, col4 = st.columns(4)
        col1.metric("Competitors Nearby", len(nearby_df))
        col2.metric("Market Capture (Huff)", f"{huff_prob * 100:.1f}%")
        col3.metric("Density (per km²)", f"{market_density:.2f}")
        col4.metric("Commercial Hotspot", "✅ Yes" if is_hotspot else "⚠️ No", help=f"{total_nearby_businesses} businesses within 0.5km")

        st.divider()

        colA, colB = st.columns(2)
        with colA:
            st.markdown("### ⚪ White Space Analysis (Monopoly Risk)")
            if d_min == float('inf'):
                st.success("✅ **Pristine White Space:** No competitors found. Perfect local monopoly.")
            else:
                status = "✅ Strong (Monopoly Potential)" if d_min >= (radius * 0.5) else "⚠️ Weak (Highly Competitive)"
                st.write(f"- **Nearest Direct Competitor:** {d_min:.3f} km")
                st.write(f"- **Status:** {status}")

        with colB:
            st.markdown("### 🧲 Logistics & Footfall Indicators")
            for pole, dist in pole_distances.items():
                st.write(f"- **Distance to {pole}:** {dist:.2f} km")

    # --- TAB 2: ANALYTICS & VISUALS ---
    with tab2:
        st.subheader("Market Distribution Analytics")
        if len(all_matches_df) > 0:
            col_chart1, col_chart2 = st.columns(2)

            with col_chart1:
                fig_hist = px.histogram(
                    all_matches_df, x="Distance_km", nbins=20,
                    title="Competitor Distance Distribution",
                    labels={"Distance_km": "Distance from target (km)"},
                    color_discrete_sequence=['#4CAF50']
                )
                fig_hist.add_vline(x=radius, line_dash="dash", line_color="red", annotation_text=f"Search Radius ({radius}km)")
                st.plotly_chart(fig_hist, use_container_width=True)

            with col_chart2:
                fig_scatter = px.scatter(
                    all_matches_df.head(50), x="Distance_km", y="AI_Confidence",
                    size="AI_Confidence", color="AI_Confidence", hover_name="EnterpriseName",
                    title="Threat Matrix: Proximity vs. Relevance",
                    labels={"Distance_km": "Distance (km)", "AI_Confidence": "Match Confidence (%)"},
                    color_continuous_scale="Reds"
                )
                fig_scatter.add_vline(x=radius, line_dash="dash", line_color="gray")
                st.plotly_chart(fig_scatter, use_container_width=True)

            if len(nearby_df) > 0:
                st.markdown("### Top Competitors in Radius")
                top_nearby = nearby_df.sort_values('Distance_km').head(10)
                fig_bar = px.bar(
                    top_nearby, x="Distance_km", y="EnterpriseName", orientation='h',
                    color="AI_Confidence", title=f"Closest Competitors (Top 10 within {radius}km)",
                    labels={"Distance_km": "Distance (km)", "EnterpriseName": "Business Name"},
                    color_continuous_scale="Blues"
                )
                fig_bar.update_layout(yaxis={'categoryorder':'total descending'})
                st.plotly_chart(fig_bar, use_container_width=True)
        else:
            st.warning("Not enough data to generate visualizations. No competitors found.")

    # --- TAB 3: RAW DATA ---
    with tab3:
        st.subheader(f"🗺️ Competitors within {radius}km ({len(nearby_df)} total)")
        if len(nearby_df) > 0:
            show_df = nearby_df[['EnterpriseName', 'Activities', 'AI_Confidence', 'Distance_km']].sort_values('Distance_km').copy()
            show_df['Distance_km'] = show_df['Distance_km'].round(3)
            st.dataframe(show_df, use_container_width=True)
        else:
            st.info("No competitors found in this radius.")

        st.subheader(f"🌐 All Global Matches in Database ({len(all_matches_df)} total)")
        if len(all_matches_df) > 0:
            all_show = all_matches_df[['EnterpriseName', 'Activities', 'AI_Confidence', 'Distance_km']].sort_values(['AI_Confidence', 'Distance_km'], ascending=[False, True]).copy()
            all_show['Distance_km'] = all_show['Distance_km'].round(3)
            st.dataframe(all_show, use_container_width=True)
```
